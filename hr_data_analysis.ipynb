#### (3) Headings -- Customizing HR dataset headings.

```python
print(hr_data.columns)
```

```python
# Updating column names that will be used later for analysis
hr_data.rename(columns= {'PerfScoreID': 'Performance_Score (0-5)'}, inplace = True)
hr_data.rename(columns= {'EngagementSurvey': 'Engagement_Level (0-5)'}, inplace = True)
hr_data.rename(columns= {'EmpSatisfaction': 'EE_Satisfaction_Level (0-5)'}, inplace = True)
```

```python
# Converting specific columns to strings so they are not considered 'numerical'
hr_data[['EmpID', 'MarriedID', 'MaritalStatusID', 'GenderID', 'EmpStatusID', 'DeptID', 'FromDiversityJobFairID', 'Termd', 'PositionID', 'Zip', 'ManagerID']] = hr_data[['EmpID', 'MarriedID', 'MaritalStatusID', 'GenderID', 'EmpStatusID', 'DeptID', 'FromDiversityJobFairID', 'Termd', 'PositionID', 'Zip', 'ManagerID']].astype(str)
```

#### (4) Describe -- Using DataFrame describe() function to calculate descriptive stats on HR dataset.

```python
hr_data.describe() # general descriptive statistics for all numerical values in HR dataset
```

A couple highlights on descriptive statistics from numerical values in the HR dataset:

(1) Salary:
-  Average = $69,020.68
-  Median = $62,810.00

Median salary is a more realistic salary for most employees at this company due the median being significantly lower ($6,210.68) than the average. In other words, there are some outliers on the top end of the salaries at this company that skew the average salary.

(2) Engagement:
-  Average: 4.11 out of 5.00
-  Median: 4.28 out of 5.00
-  Standard Deviation: 0.79

Employee engagement looks relatively great at this company. Both average and median are close together and both over 4.00 on a 5.00 scale. Also, standard deviation is low at 0.79 which tells me the extremes are kept in check (data isn't super spread out).
